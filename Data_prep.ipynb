```python
df = pd.read_csv("/content/drive/MyDrive/Licencjat/tweets.csv", encoding='utf8')
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("visualizing_missing_data_with_heatmap_Seaborn_Python.png", dpi=100)
```

```python
na = df.isna().sum()
na
```

```python
df = df[df['text'].notna()]
df.shape
```

Kolumna "Text" zawiera 2 675 037 brakujcych wartoci (0.78%), po usuniciu zostao nam 737 774 obserwacje

```python
df_duplicated = df[df.duplicated()]
df_duplicated.sort_values("text")
```

```python
df = df.drop_duplicates()
```

```python
display(df.shape)
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.displot(
    data= df.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25
)
plt.savefig("visualizing_missing_data_with_barplot_Seaborn_distplot.png", dpi=100)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("visualizing_missing_data_with_heatmap_Seaborn_Python.png", dpi=100)
```
